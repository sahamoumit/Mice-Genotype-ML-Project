#5. Using Five fold cross validation and performing settings for the hyperparameters for Binary classifier

```python
X_train= df2.values
Y_train = df['Genotype'].values
Y_train2 = df['Treatment_Behavior'].values
```

```python
X_test= df_test2.values
```

```python
X_train1= pd.DataFrame(X_train)
```

```python
from numpy import arange
from pandas import read_csv
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import ElasticNet
from sklearn import metrics
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.feature_selection import RFECV
```

# 5.A: Elastic net for Binary classification

```python
search = LogisticRegressionCV(cv=5, penalty='elasticnet', l1_ratios=[0.5], solver = 'saga', tol=0.01, random_state = 0)
```

```python
rfe_reg = RFECV(search, step=1)
rfe_reg_op = rfe_reg.fit(X_train, Y_train)
```

```python
print('Accuracy for Logistic Regression:', metrics.accuracy_score(Y_train, rfe_reg_op.predict(X_train)))
```

# No. of Imp. Features

```python
print('number of important features are: ', len(rfe_reg_op.get_feature_names_out(input_features = X_train1.columns)))
rfe_reg_op.get_feature_names_out(input_features = X_train1.columns)
```

# Predictions on test data

```python
from sklearn.metrics import classification_report, confusion_matrix 
```

```python
Y_pred_L = rfe_reg_op.predict(X_test)
print('accuracy:', metrics.accuracy_score(Y_test, Y_pred_L))
print(confusion_matrix(Y_test, Y_pred_L))
print(classification_report(Y_test, Y_pred_L))
```

#5.d:Random Forest Binary classifcation

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
forest_params = [{'n_estimators': list(range(1, 20)), 'max_depth': list(range(10,15)), 'max_features': list(range(0,20))}]

clf = GridSearchCV(rfc, forest_params, cv = 5, scoring='accuracy')

clf.fit(X_train, Y_train)
```

```python
print(clf.best_params_)

print(clf.best_score_)
```

```python
def RFECV_ranking(X_train,Y_train):
  rf = RandomForestClassifier(n_estimators= 150, random_state=300)
  rfecv = RFECV(estimator=rf, step=1, cv=5, scoring='accuracy')
  rfecv.fit(X_train,Y_train)
  print('Optimal number of features: {}'.format(rfecv.n_features_))

  features = [f for f,s in zip(X_train1.columns, rfecv.support_) if s]
  return features
```
